```python
import pandas as pd

df = pd.read_csv(r"C:\Users\Kota India\Downloads\student\student-mat.csv", sep=";")

print(df.head())
```

```python
print("Rows, Columns:", df.shape)
print("\nColumns:")
print(df.columns)
```

```python
# Check missing values

print(df.isnull().sum())
```

```python
# Check duplicate rows

print("Duplicate Rows:", df.duplicated().sum())
```

```python
average_grade = df["G3"].mean()

print("Average Final Grade:", average_grade)
```

```python
above_15 = df[df["G3"] > 15]

print("Students scoring above 15:", len(above_15))
```

```python
correlation = df["studytime"].corr(df["G3"])

print("Correlation between Study Time and Final Grade:", correlation)
```

```python
gender_avg = df.groupby("sex")["G3"].mean()

print(gender_avg)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.hist(df["G3"], bins=10)

plt.title("Distribution of Final Grades")
plt.xlabel("Final Grade")
plt.ylabel("Number of Students")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(df["studytime"], df["G3"])

plt.title("Study Time vs Final Grade")
plt.xlabel("Study Time")
plt.ylabel("Final Grade")

plt.show()
```

```python
gender_avg.plot(kind="bar")

plt.title("Average Final Grade by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Grade")

plt.show()
```
